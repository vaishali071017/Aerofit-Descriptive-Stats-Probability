## **Remove/clip the data between the 5 percentile and 95 percentile**

```python
lower_percentile=data.quantile(0.05)
upper_percentile=data.quantile(0.95)
```

```python
clipped_data = data.clip(lower=lower_percentile, upper=upper_percentile, axis=1)
clipped_data
```

```python
clipped_data.drop(['Product', 'Gender', 'MaritalStatus'], axis=1, inplace=True)
clipped_data
```

```python
category_data = data[['Product', 'Gender', 'MaritalStatus']]
category_data
```

```python
df = pd.concat([clipped_data, category_data], axis=1)
df
```

# **4. Representing the probability**
## **Marginal Probability**

```python
pd.concat( [
    df.Product.value_counts(),
    df.Product.value_counts(normalize=True)
    ],
    keys=['counts', 'Marginal_Prob'], axis=1
)
```

## **Probability that the customer buys a product based on each column.**

```python
def product_given_gender(gender, print_marginal=False):
  df1=pd.crosstab(index=df['Gender'], columns=[df['Product']])
  p_781= df1['KP781'][gender] / df1.loc[gender].sum()
  p_481= df1['KP481'][gender] / df1.loc[gender].sum()
  p_281= df1['KP281'][gender] / df1.loc[gender].sum()

  if print_marginal:
    print(f"P(Male): {df1.loc['Male'].sum()/len(df):.2f}")
    print(f"P(Female): {df1.loc['Female'].sum()/len(df):.2f}")

  print(f"P(KP781/{gender}):{p_781:.2f}")
  print(f"P(KP481/{gender}):{p_481:.2f}")
  print(f"P(KP281/{gender}):{p_281:.2f}")

product_given_gender('Male',True)
product_given_gender('Female')
```

```python
def product_given_maritalstatus(status, print_marginal=False):
  df1=pd.crosstab(index=df['MaritalStatus'], columns=[df['Product']])
  p_781= df1['KP781'][status] / df1.loc[status].sum()
  p_481= df1['KP481'][status] / df1.loc[status].sum()
  p_281= df1['KP281'][status] / df1.loc[status].sum()

  if print_marginal:
    print(f"P(Single): {df1.loc['Single'].sum()/len(df):.2f}")
    print(f"P(Partnered): {df1.loc['Partnered'].sum()/len(df):.2f}")

  print(f"P(KP781/{status}):{p_781:.2f}")
  print(f"P(KP481/{status}):{p_481:.2f}")
  print(f"P(KP281/{status}):{p_281:.2f}")

product_given_maritalstatus('Single',True)
product_given_maritalstatus('Partnered')
```
